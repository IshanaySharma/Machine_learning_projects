```python
#Image Filters
f1 = cv2.imread('/content/tiger2.jpg')
f1 = cv2.cvtColor(f1,cv2.COLOR_BGR2RGB) #Coverting default bgr img to rgb
plt.imshow(f1)
plt.show()
```

```python
#Creating different kernels
k1 = np.ones((5,5))/25
k2 = np.ones((9,9))/81
k3 = np.ones((11,11))/121

#Applying 'Blurred' filter
res1 = cv2.filter2D(f1,-1,k1)
res2 = cv2.filter2D(f1,-1,k2)
res3 = cv2.filter2D(f1,-1,k3)

plt.imshow(f1),plt.title('Original'),plt.show()
plt.imshow(res1),plt.title('Blurred with 5*5 kernel'),plt.show()
plt.imshow(res2),plt.title('Blurred with 9*9 kernel'),plt.show()
plt.imshow(res3),plt.title('Blurred with 11*11 kernel'),plt.show()
```

```python
#Custom filters
k4 = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]]) #Sharpening kernel
k5 = np.array([[0,-1,-1],[1,0,-1],[1,1,1]]) #Emboss kernel
```

```python
g = cv2.imread('/content/panda.jpg')
g = cv2.cvtColor(g,cv2.COLOR_BGR2RGB) #Coverting default bgr img to rgb
plt.imshow(g)
plt.show()
```

```python
r1 = cv2.filter2D(g,-1,k4)
r2 = cv2.filter2D(g,-1,k5)

plt.imshow(g),plt.title('Original'),plt.show()
plt.imshow(r1),plt.title('Sharpen kernel'),plt.show()
plt.imshow(r2),plt.title('Emboss kernel'),plt.show()
```
